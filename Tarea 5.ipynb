### 1 Maximice $z = x_1 + x_2$

Sujeto a:
$$ x_1 + 2x_2 \leqslant 10 $$
$$ 2x_1 + x_2 \leqslant 16 $$
$$ -x_1 + x_2 \leqslant 3 $$
$$ x_1 \geqslant 0, \quad x_2 \geqslant 0$$

```python
import numpy as np #Mejor matemática
import matplotlib.pyplot as plt #Gráficos

#Puntos de x1 a graficar
x1 = np.arange(0,10.1,0.1)
#Hasta 10 debido a la primera ecuación

# x1 + 2x2 <= 10
def F1(x1):
    x2 = (10 - x1)/2
    return x2
#Graficamos la función 1 en todos los puntos x1, rellenando entre 0 y la función
#Si fuera >0 rellenaría entre un punto máximo y la función, como pasa en el siguiente ejercicio
plt.fill_between(x1, 0, F1(x1), label = r'$ x_1 + 2x_2 \leq 10 $', alpha = 0.3)
#Label es la etiqueta para el gráfico y alpha la transparencia
#Hacemos lo mismo con las otras funciones

# 2x1 + x2 <= 16
def F2(x1):
    x2 = 16 - 2*x1
    return x2
plt.fill_between(x1[:80],0,F2(x1)[:80], label = r'$ 2x_1 + x_2 \leq 16 $', alpha = 0.3)
#En este caso graficamos solo los primeros 80 valores ya que después se hace negativo
#Es decir las restricciones finales --que sean mayores a 0-- está implícita en el gráfico

#-x1 + x2 <= 3
def F3(x1):
    x2 = 3 + x1
    return x2
plt.fill_between(x1,0,F3(x1), label = r'$ -x_1 + x_2 \leq 3 $', alpha = 0.3)

#Para conveniencia también graficamos las esquinas
#Las intersecciones entre ecuaciones son obtenidas más adelante, esto es para ver en el gráfico
px1 = [8,0,22/3,4/3]
px2 = [0, 3, 4/3, 13/3]
plt.plot(px1,px2, '*', label = 'Esquinas')

plt.title(r'z = x_1 + x_2')
plt.legend(loc = 'upper center') #Muestra los 'labels' en el gráfico
plt.xlabel('x1') #Etiquetar el eje x
plt.ylabel('x2') # Etiquetar el eje y
plt.show()
```

Los candidatos a puntos máximos dadas las restricciones son el punto de F2 en que $x_2 = 0$, el punto de F3 en que $x_1 = 0$ y las intersecciones entre F1 y F2 y entre F1 y F3, es decir los puntos $(8,0)$, $(0,3)$ y las soluciones a el sistema:

$$ x_1 + 2x_2 = 10 $$
$$ 2x_1 + x_2 = 16 $$

con solución $x_1 = 22/3$ y $x_2 = 4/3$, el sistema:
$$ x_1 + 2x_2 = 10 $$
$$ -x_1 + x_2 = 3 $$
con solución $x_1 = 4/3$ y $x_2 = 13/3$

Los puntos $(8,0)$ y $(0,3)$ se evaluan a $z = 8$ y $z = 3$ respectivamente.
La primera solución de sistemas evalua a $z = 26/3$ mientras que la segunda evalua a $z = 17/3$ por lo que la primera solución es la correcta $(22/3,4/3)$.

### 2 Minimice $z = x_1 + 3x_2$

Sujeto a:
$$ x_1 + 2x_2 \geqslant 6 $$
$$ x_1 - x_2 \leqslant 3 $$
$$ x_1 \geqslant 0, \quad x_2 \geqslant 0$$

```python
x1 = np.arange(0,10.1,.1) 
#En este caso el gráfico nunca cubrirá toda el área de soluciones posibles
#Pero con este tamaño se ven bien los puntos que interesan

#x1 + 2x2 >= 6
def F1(x1):
    x2 = (6 - x1)/2
    return x2
plt.fill_between(x1, 10, F1(x1), label = r'$ x_1 + 2x_2 \geq 6 $', alpha = 0.3)

#x1 - x2 <= 3
def F2(x1):
    x2 = -3 + x1 #Notese el cambio de signo para el gráfico, ahora es mayor o igual, lo que afecta el gráfico
    return x2
plt.fill_between(x1, 10, F2(x1), label = r'$ x_1 - x_2 \leq 3 $', alpha = 0.3)

#Esquinas
px1 = [4,0]
px2 = [1,3]
plt.plot(px1,px2, '*', label = 'Esquinas')

plt.title(r'z = x_1 + 3x_2')
plt.legend(loc = 'lower center')
plt.xlabel('x1')
plt.ylabel('x2') 
plt.show()
```

En este caso solo hay que hacer una intersección entre las dos ecuaciones, lo que da $x_1 = 4$ y $x_2 = 1$, la otra esquina esta dadas por F1 cuando $x_1 = 0$, es decir $(0,3)$.
Estos puntos evaluan a $z = 7$ y $z = 9$ respectivamente, por lo que la solución es $(4,1)$.


### 3 Minimice $z = 2x_1 - x_2$

Sujeto a:
$$ x_1 + 2x_2 \geqslant 6 $$
$$ x_1 - x_2 \leqslant 3 $$
$$ x_1 \geqslant 0, \quad x_2 \geqslant 0$$

En este caso claramente el mínimo estaría donde $x_1$ es menor y $x_2$ es mayor, pero ninguna de las dos condiciones limitan el crecimiento de $x_2$, es decir la solución sería algo como $(0,\infty)$

Pero aprovecharé el espacio para mostrar una forma alternativa (mejor?) de graficar el problema:

```python
plt.figure(figsize = (12,8))

x1 = np.arange(0,10.1,.15) 

plt.plot(x1, F1(x1), lw = 3, label = r'$ x_1 + 2x_2 \geq 6 $')
plt.plot(x1, F2(x1), lw = 3, label = r'$ x_1 - x_2 \leq 3 $')

x2 = np.arange(0,F2(x1[-1]) + .1,.1) 

#Función a minimizar
def z(x1,x2):
    return 2*x1 - x2

#Encontrar todos los pares que cumplan las condiciones
pares = [(p1, p2) #Par
         for p1 in x1 #Iterar en x1
         for p2 in x2 #Iterar en x2
         if p1 + 2*p2 >= 6 #Contar el par si cumple la condición 1
         and p1 - p2 <= 3 ] #Y cumple la condición 2
pares = np.asarray(pares) #Convertir a un array, lo anterior era algo para listas
valor = z(pares[:,0],pares[:,1]) #Evaluar los pares en la función

# Graficamos los puntos que caen dentro de las condiciones
plt.scatter(pares[:,0],pares[:,1], c = valor, cmap = 'jet', label = 'z')


cb = plt.colorbar()
cb.set_label('z')
plt.title(r'z = 2x_1 - x_2')
plt.legend(loc = 'lower center')
plt.xlabel('x1')
plt.ylabel('x2') 
plt.show()
```

### 4
#### Una fábrica hace 3 componentes, A, B y C usando el mismo proceso de producción para cada uno.

Una unidad de A toma 1 hora, una unidad de B toma 0.75 horas y una unidad de C toma 0.5 horas.
Además, C se termina a mano, un actividad que necesita 0.25 horas por unidad.

Cada semana el tiempo total de producción no debe exceder de 300 horas de máquinas, y 45 horas manuales.

Los componentes son ensamblados para hacer DOS productos.
Un producto P1 consta de 1 unidad de A y 1 unidad de C que se venden por 30 dólares - mientras que el otro producto P2 consiste en 2 unidades de B y 1 unidad de C y se vende por 45 dolares.

Se venden 130 de P1 y 100 de P2 semanal

Formula el problema de planificar la producción semanal para maximizar los ingresos totales.

La función a optimizar es:
$$ Ganancias = 30P_1 + 45P_2 $$
Teniendo las siguientes igualdades:
$$ P_1 = A + C $$
$$ P_2 = 2B + C $$

Además la producción está sujeta a las siguientes restricciones:
$$ A + 0.75B + 0.5C \leqslant 300$$
$$ 0.25 C \leqslant 45 $$

Y debido a la cantidad de ventas, no sirve producir P1 y P2 más de un cierto punto, es decir:
$$ P_1 \leqslant 130 $$
$$ P_2 \leqslant 100 $$

La ganancia uno la puede mirar como:
$$ Ganancias = 30(A + C)+ 45(2B + C) $$
$$ Ganancias = 30A + 90B + 75C $$

Pero no hay que ignorar que cantidades de A, B y C que no puedan formar un $P$ entero no aportan a las ganacias, es por esto que en vez de convertir las restricciones de productos a materiales, encuentro mas conveniente convertir las restricciones de materiales a productos.

Así tenemos que la restricción de horas de maquinaria se convierte a:

$$ P_1 + 1.5P_2 + 0.5(P_1 + P_2)  \leqslant 300 $$
$$ P_1 + 1.5P_2 + 0.5P_1 + 0.5P_2 \leqslant 300 $$
$$ 1.5P_1 + 2P_2 \leqslant 300 $$

Y la restricción de horas manuales se convierte en:

$$ 0.25 (P_1 + P_2) \leqslant 45 $$
$$ P_1 + P_2 \leqslant 180 $$

Entonces podemos graficar de igual manera que en otros problemas.

```python
plt.figure(figsize = (12,8))

P1 = np.arange(0,130.1,1) #P1 <= 130

P2 = np.arange(0,100.1,1) #P2 <= 100
plt.plot(P1,[100]*len(P1), lw = 3, label = r'$P_2 <= 100$')
#Ganancias
def ganancias(p1,p2):
    return 30*p1 + 45*p2

# 1.5P1 + 2P2 <= 300
def F1(p1):
    p2 = (300 - 1.5*p1)/2
    return p2
plt.plot(P1,F1(P1), label = r'$ 1.5P_1 + 2P_2 \leqslant 300 $', lw = 3)

# P1 + P2 <= 180
def F2(p1):
    p2 = 180 - p1
    return p2
plt.plot(P1,F2(P1), label = r'$ P_1 + P_2 \leqslant 180 $', lw = 3)

#Encontrar todos los pares que cumplan las condiciones
pares = [(p1, p2)
         for p1 in P1
         for p2 in P2 
         if 1.5*p1 + 2*p2 <= 300
         and p1 + p2 <= 180 ] 

pares = np.asarray(pares) #Convertir a un array, lo anterior era algo para listas
valor = ganancias(pares[:,0],pares[:,1]) #Evaluar los pares en la función
plt.scatter(pares[:,0],pares[:,1], c = valor, cmap = 'jet', label = 'Ganancias')
cb = plt.colorbar()
cb.set_label('Ganancias')

plt.title(r'Problema 4: Ganacias = $30P_1 + 45P_2$')
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2') 
plt.show()
```

De los datos usados para graficar se podría obtener la solución numérica, pero la solución analítica considerando las intersecciones como en los primeros ejercicios.

El primer candidato es el punto $(0,100)$, que se ve claramente que no es, pues tiene ganancias de solo $3000.

El segundo candidato es el punto de F1 (naranja) en que $P_2 = 100$, es decir $(66.6,100)$. 
Si consideramos el punto entero más cercano $(66,100)$ en este punto las ganancias son de $6480.

El tercer candidato es la intersección entre F1 y F2 (Verde), dado por:
$$ P_1 + P_2 = 180 $$
$$ 1.5P_1 + 2P_2 = 300 $$

Donde la solución es el punto $(120,60)$, que se evalua a ganancias de $6300, menores que el punto anterior.

El último candidato  es el punto de la función F2 donde $P_1 = 130$, es decir $(130, 50)$.
Este punto daría ganancias de $6150.

La solución entonces es $P_1 = 66$ y $P_2 = 100$.
